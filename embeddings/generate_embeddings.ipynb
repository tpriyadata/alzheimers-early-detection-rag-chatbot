### Load the chunks

```python
# Cell 1: Load the chunks we created in Block 2
# These are the text pieces we will convert to vectors

import json

with open("../data/processed/chunks.json", "r") as f:
    chunks = json.load(f)

print(f"Loaded {len(chunks)} chunks")
print("\nFirst chunk text:")
print(chunks[0]["text"][:200])
```

### Load The Embedding Model

```python
# Cell 2: Load the embedding model
# all-MiniLM-L6-v2 converts text to 384-dimension vectors
# It's small, fast and works great for medical text

from sentence_transformers import SentenceTransformer

print("Loading embedding model...")

model = SentenceTransformer("all-MiniLM-L6-v2")

print("Model loaded successfully!")
print(f"Model name: all-MiniLM-L6-v2")
print(f"Vector size it produces: 384 dimensions")
```

### Generate Embeddings

```python
# Cell 3: Convert all chunks to vectors
# This is the core of the embedding process

# Extract just the text from each chunk
texts = [chunk["text"] for chunk in chunks]

print(f"Embedding {len(texts)} chunks...")

# Convert all texts to vectors at once
embeddings = model.encode(texts, show_progress_bar=True)

print(f"\nDone!")
print(f"Embeddings shape: {embeddings.shape}")
print(f"Each chunk → {embeddings.shape[1]} numbers")
```

###  Save The Embeddings

```python
# Cell 4: Save embeddings to disk
# We save both the vectors AND the original chunks
# WHY? We need text later to show the user

import numpy as np
import os

os.makedirs("../data/processed", exist_ok=True)

# Save vectors as numpy file
np.save("../data/processed/embeddings.npy", embeddings)

# Save chunk texts separately
with open("../data/processed/chunks.json", "w") as f:
    json.dump(chunks, f, indent=2)

print(f"Saved {len(embeddings)} embedding vectors")
print(f"Shape: {embeddings.shape}")
print(f"\nFiles saved:")
print(f"  → data/processed/embeddings.npy")
print(f"  → data/processed/chunks.json")
```
